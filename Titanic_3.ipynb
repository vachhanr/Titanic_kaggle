```python
print("Before", train.shape, test.shape, combine[0].shape, combine[1].shape)
train  =  train.drop(columns = ["Ticket","Cabin"])
test = test.drop(columns = ["Ticket","Cabin"])
combine = [train,test]
print("After", train.shape, test.shape, combine[0].shape, combine[1].shape)
```

```python
for dataset in combine:
    dataset["Title"] = dataset.Name.str.extract('([A-Za-z]+)\.',expand = False)

pd.crosstab(train['Title'],train['Sex'])
```

```python
for dataset in combine:
    dataset["Title"] = dataset["Title"].replace(['Lady','Countess','Capt', 'Col','Don', 'Dr', 'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona'], 'Rare')
    dataset["Title"] = dataset["Title"].replace("Mlle","Miss")
    dataset["Title"] = dataset["Title"].replace("Mme","Mrs")
    dataset["Title"] = dataset["Title"].replace("Ms","Miss")
    
train[["Title","Survived"]].groupby(["Title"], as_index=False).mean()
```

```python
for dataset in combine:
    dataset["Sex"] = dataset["Sex"].map({"female":1,"male":0}).astype(int)

train.head()
```

```python
grid = sns.FacetGrid(train, row = "Pclass", col= "Sex", aspect = 1.6, size = 2.2)
grid.map(plt.hist, "Age",alpha = 0.5, bins = 20)
grid.add_legend()
```

```python
guess_ages = np.zeros((2,3))
guess_ages
```

```python
for dataset in combine:
    for i in range(0,2):
        for j in range(0,3):
            guess_df = dataset[(dataset["Sex"] == i) & (dataset["Pclass"] == j+1)]["Age"].dropna()
            age_guess = guess_df.median()
            guess_ages[i,j] = int(age_guess/0.5 + 0.5) * 0.5
            
    for i in range(0,2):
        for j in range(0,3):
            dataset.loc[(dataset.Age.isnull()) & (dataset.Sex == i) & (dataset.Pclass == j+1),"Age"] = guess_ages[i,j]
    dataset["Age"] = dataset["Age"].astype(int)
            
train.head()
            
```

```python
train["AgeBand"] = pd.cut(train["Age"],5)
train[["AgeBand","Survived"]].groupby(by = ["AgeBand"],as_index=False).mean().sort_values(by = ["AgeBand"],ascending = True)
```

```python
for dataset in combine:
    dataset.loc[dataset["Age"]<=16,"Age"] = 0
    dataset.loc[(dataset["Age"]>16) & (dataset["Age"]<=32),"Age"]=1
    dataset.loc[(dataset["Age"]>32) & (dataset["Age"]<=48),"Age"]=2
    dataset.loc[(dataset["Age"]>48) & (dataset["Age"]<=64),"Age"]=3
    dataset.loc[dataset["Age"]>64,"Age"] 
    
train.head()
```

```python
train = train.drop(columns = ["AgeBand"])
combine = [train,test]
train.head()
```

```python
for dataset in combine:
    dataset["FamilySize"] = dataset["SibSp"] + dataset["Parch"] + 1

train[["FamilySize","Survived"]].groupby(["FamilySize"],as_index=False).mean().sort_values("Survived",ascending = False)
```

```python
for dataset in combine:
    dataset["IsAlone"] = 0
    dataset.loc[dataset["FamilySize"] == 1, "IsAlone"] = 1
    
train[["Survived","IsAlone"]].groupby(["IsAlone"]).mean()
```

```python
train = train.drop(columns = ["FamilySize","SibSp","Parch"])
test = test.drop(columns = ["FamilySize","SibSp","Parch"])
combine = [train,test]
train.head()
```

```python
for dataset in combine:
    dataset["Age*Pclass"] = dataset.Age * dataset.Pclass
    
train.loc[:,['Age*Pclass','Age','Pclass']].head(10)
```

```python
freq_port = train.Embarked.dropna().mode()[0]
freq_port
```

```python
for dataset in combine:
    dataset["Embarked"] = dataset['Embarked'].fillna(freq_port)
    
train[["Embarked","Survived"]].groupby(["Embarked"],as_index=False).mean().sort_values(by = "Survived", ascending = False)
```

```python
for dataset in combine:
    dataset["Embarked"] = dataset["Embarked"].map({"C":1, "Q": 2, "S":0}).astype(int)
    
train.head()
```
